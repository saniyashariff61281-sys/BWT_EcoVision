```python
import face_recognition
import os
import cv2
import numpy as np

known_encodings = []
known_names = []

dataset_path = "students"

for file in os.listdir(dataset_path):

    # Skip folders
    if file.startswith("."):
        continue

    img_path = os.path.join(dataset_path, file)

    # Make sure it's a file (not folder)
    if os.path.isfile(img_path):
        image = face_recognition.load_image_file(img_path)
        encodings = face_recognition.face_encodings(image)

        if len(encodings) > 0:
            known_encodings.append(encodings[0])
            name = os.path.splitext(file)[0]
            known_names.append(name)

print("Encoding Completed Successfully!")
```

```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import face_recognition

# Load class image
test_image = face_recognition.load_image_file("class.jpeg")
test_image_cv = cv2.cvtColor(test_image, cv2.COLOR_RGB2BGR)

# Detect faces
face_locations = face_recognition.face_locations(test_image)
face_encodings = face_recognition.face_encodings(test_image, face_locations)

attendance = []

for face_encoding in face_encodings:
    matches = face_recognition.compare_faces(known_encodings, face_encoding)
    name = "Unknown"

    face_distances = face_recognition.face_distance(known_encodings, face_encoding)

    if len(face_distances) > 0:
        best_match_index = np.argmin(face_distances)
        if matches[best_match_index]:
            name = known_names[best_match_index]

    attendance.append(name)

print("Attendance List:", attendance)
```

```python
import csv
from datetime import datetime

now = datetime.now()
date_string = now.strftime("%Y-%m-%d")
time_string = now.strftime("%H:%M:%S")

with open("attendance.csv", "a", newline="") as f:
    writer = csv.writer(f)
    for name in attendance:
        if name != "Unknown":
            writer.writerow([name, date_string, time_string])

print("Attendance Saved Successfully!")
```
